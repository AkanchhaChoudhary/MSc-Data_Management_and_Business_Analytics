```python
folder = './1.5_mnist_images/'

filenames = [folder+f for f in os.listdir(folder)]
# filenames = [f for f in filenames if '.jpg' in f]
filenames
```

Lets create a split of 80-20 based on the names

```python
train_files, test_files = train_test_split(filenames, test_size = 0.2, random_state = 123)

test_files[0:4]
```

```python
import numpy as np

labels = [int(name.split('_')[-1].split('.')[0]) for name in filenames] # This extracts the digital label from the 

#using only numpy

values = list(set(labels))
arr_labels = np.array(labels)

counts = {}
for v in values:

    counts[v] = sum(arr_labels == v)

    for key, value in counts.items():
        print(f"{key}: {value}")
```

```python
from collections import Counter
label_counts = Counter(labels)
label_counts
```

```python
sorted_label_counts = {k: label_counts[k] for k in sorted(label_counts)}
sorted_label_counts
```
