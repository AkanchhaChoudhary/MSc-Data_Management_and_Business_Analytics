```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
m=.8
x=np.random.rand(20)
x2=np.random.rand(20)+1
y = np.exp(1.4*x)-np.exp(m*x**2)
y2=np.exp(1.4*x2)-np.exp(m*x2**2)
plt.plot(x,y, '*')
plt.plot(x2,y2, 'g*')
plt.xlim(0,2)
plt.ylim(0,3.5)
plt.show()
```
