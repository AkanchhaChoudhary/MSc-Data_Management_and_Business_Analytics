```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [6, 3] # this line ensures that now onwards thesize of lot will remain fixed to 6 and 3

np.random.seed(10)
```

```python
X = np.random.randn(10000) #this will generate a lot of random number from normal distribution "np.random.randn(10000)" is use o generate random number and here 10000 such number will be generated and then multioying with 3 menas putting into std of 3 and 10 means all random will be shifted to the right by 10
plt.hist(X, 30)
plt.show()
print(f'mean is {np.mean(X)} and sigma is {np.std(X)}')
```

Let's generate some random normal data:


```python
X= 10 + 3*np.random.randn(10000) #this will generate a lot of randomnumuner from normal distribution "np.random.randn(10000)" is use o generate random number and here 10000 such number will be generated and then multioying with 3 menas putting into std of 3 and 10 means all random will be shifted to the right by 10
plt.hist(X, 30)
plt.show()
print(f'mean is {np.mean(X)} and sigma is {np.std(X)}')
```

```python
X.shape
```

Now, let's  take different samples (for instance 1000) of size N=100.

```python
N = 100  #  size of each sample
K = 1000 # number of samples
sample_mean = np.zeros(K)
sample_std = np.zeros(K)

for i in range(K):
    s = np.random.choice(X, N)
    m=np.mean(s)
    sigma=np.std(s)
    sample_mean[i]=m
    sample_std[i]=sigma
```

```python
print(f'mean is {np.mean(sample_mean)} and sigma is {np.std(sample_mean)}')
plt.hist(sample_mean, 30)
plt.show()
```

According to the theory, the average of the sample mean must be $\mu = 10$, and the average standard error is $\sigma/\sqrt{N} = 3/\sqrt{100}=0.3$

The empirical values are quite close.

##  Let's try now with a non normal distribution (exponential distribution)

```python
W = np.random.exponential(scale=1, size=10000)
mean_W = np.mean(W)
std_W = np.std(W)
plt.hist(W, 30)
plt.axvline(mean_W-std_W, color='orange')
plt.axvline(mean_W, color='green')
plt.axvline(mean_W+std_W, color='orange')

print(f'mean is {mean_W} and sigma is {std_W}')
plt.show()
```

We repeat the same code as before.


```python
N = 100 # size of each sample
K = 1000 # number of samples
sample_mean= np.zeros(K)
sample_std= np.zeros(K)

for i in range(K):
    s = np.random.choice(W, N)
    m=np.mean(s)
    sigma=np.std(s)
    sample_mean[i]=m
    sample_std[i]=sigma
```

```python
m = np.mean(sample_mean)
s = np.std(sample_mean)
print(f'mean is {m} and sigma is {s}')
plt.hist(sample_mean, 30, density=True)
plt.show()
```

According to the theory, the average of the sample mean must be $\mu \sim 1$, and the average standard error is $s = \sigma/\sqrt{N} \sim 1/\sqrt{100}=0.1$

I can also plot on top of the histogram the theoretical normal curve with the expected mean and standard deviation.

```python
m = np.mean(sample_mean)
s = np.std(sample_mean)
z = np.linspace(m-4*s, m+4*s, 100)  # Range of x-values
n_z = (1 / (s * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((z - m) / s)**2)

# plot histogram
plt.hist(sample_mean, 30, density = True, label='sample data')
# Plot the standard normal distribution
plt.plot(z, n_z, label = 'theoretical curve')
plt.legend()
plt.show()
```
