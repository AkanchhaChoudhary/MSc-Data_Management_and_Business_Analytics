

# Visualizing Data with Python
##  Introduction to Matplotlib
**Matplotlib** is a Python 2D and 3D plotting library 

The matplotlib package is quite extensive and the learning curve is steep. Fortunately, there is a subpackage called **pyplot** that offers a good compromise between functionality and complexity of use. Here we will focus exclusively on the matplotlib.pyplot package. 

We begin by importing the package. Standard rules dictate that the namespace to use is **plt**. We also import the numpy and pandas packages to work with the data that we will plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Let's suppose we have some raw data. For instance:


```python
data=[3,5,4,3,8,6]
# line plot
plt.plot(data)
```

```python
plt.plot(data, '*') #points plot
```

```python
d5=pd.read_excel('data_session3.xlsx', sheet_name='data5')
d5.head()
```

## Simple Plots

The simpler plots are the line plot.
For instance, assume that numbers in data1 represent a price of a stock. We want to make a plot line with all the prices.

And this is how it is done in Python.



```python
Price = d5['Values']
plt.plot(Price)
plt.show()
```

We can also specify the 'time' or 'x' axis with a different array of numbers.


```python
Price = d5['Values']
plt.plot(Price, color='green', label='Price')
plt.legend()
plt.show()
```

Other option of marker:

```python
plt.plot(Price ,'.')  ## o= symbol '.'
plt.show()
```

which is much clearer. If we want to select other colour (blue is the default in the 'bmh' template) we can use the first letter of the colour for it.

```python
plt.plot(Price,'.', color='red')  ## r=color 'red', o= symbol 'o', 
plt.show()
```

### Plotting several lines on the same graph

```python
df_sales=pd.read_csv('sales_us.txt', skiprows=3, sep=';')
print(df_sales.shape)
df_sales.head()
```

```python
plt.figure()  # create a figure
plt.plot(df_sales['Time'], df_sales['Boston'], '-', label='Boston')  # add this to the figure
plt.plot(df_sales['Time'], df_sales['Portland'], '*-', label='Portland')  # add this too to the figure
plt.legend()
plt.show()  # when you're done, plot it.
```

# Bar Charts

```python
brands = ['Ford', 'BMW', 'Toyota', 'Fiat']
sales = [12.5, 10.1, 21, 16]
plt.figure(figsize=(4,2))
plt.bar(brands, sales, facecolor='green', alpha=0.4, label="US Market")
plt.xlabel("Brand")
plt.ylabel("Sales (M$)")
plt.legend()
```

## Scatterplots

Scatterplots can be informative about joint distributions.

```python
y = np.random.standard_normal((1000, 2))
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(df_sales['Boston'], df_sales['Portland'], 'ro')
plt.grid(True)
plt.xlabel('Boston')
plt.ylabel('Portland')
plt.title('Scatter Plot')
plt.show()
```

Saving a plot to a file is easy:

```python
plt.savefig('scatter.png')
plt.savefig('scatter.jpeg')
plt.savefig('scatter.pdf')
```
