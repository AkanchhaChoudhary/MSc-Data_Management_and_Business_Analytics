1. Compute the 5-period and 10-period moving average and plot the original data series
together with the moving averages.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/home/akanchha/Desktop/EDHEC/MSc-Data_Management_and_Business_Analytics/First_Trimester/11: STATISTICS Time Series Analysis and Forecasting/data_lecture6.xlsx'
df = pd.read_excel(file_path, sheet_name='data1')

# Compute moving averages
df['5-period MA'] = df['Sales'].rolling(window=5).mean()
df['10-period MA'] = df['Sales'].rolling(window=10).mean()

# Plot the original data and moving averages
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales'], label='Original Data')
plt.plot(df['Date'], df['5-period MA'], label='5-period MA', linestyle='--')
plt.plot(df['Date'], df['10-period MA'], label='10-period MA', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Original Data and Moving Averages')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2. What type of time series is it? Additive or multiplicative?

    Ans. This seems to be a multiplicative series.

3. Split the time series in its components and make a plot with each of them separated. The data
is taken monthly and the periodicity is annual, so the period for the time splitting will be set to
12 (months).

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose



# Decompose the time series into trend, seasonality, and residual
decomposition = seasonal_decompose(df['Sales'], period=12)  # Period set to 12 for monthly data with annual seasonality

# Plot each component separately
plt.figure(figsize=(12, 8))

# Original Data
plt.subplot(411)
plt.plot(df['Sales'], label='Original Data')
plt.legend()
plt.title('Original Data')

# Trend Component
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend()
plt.title('Trend Component')

# Seasonal Component
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend()
plt.title('Seasonal Component')


# Residual Component
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

## With the data in spreadsheet=data_2:

### 4. Compute the 5-period and 10-period moving average and plot the original data series together with the moving averages.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the 'data_2' sheet of the Excel file
data = pd.read_excel("/home/akanchha/Desktop/EDHEC/MSc-Data_Management_and_Business_Analytics/First_Trimester/11: STATISTICS Time Series Analysis and Forecasting/data_lecture6.xlsx", sheet_name="data2")

# Compute 5-period and 10-period moving averages
data['5_period_MA'] = data['Monthly milk production (pounds per cow)'].rolling(window=5).mean()
data['10_period_MA'] = data['Monthly milk production (pounds per cow)'].rolling(window=10).mean()

# Plot original data series and moving averages
plt.figure(figsize=(10, 6))
plt.plot(data['Monthly milk production (pounds per cow)'], label='Original Data')
plt.plot(data['5_period_MA'], label='5-period Moving Average', linestyle='--')
plt.plot(data['10_period_MA'], label='10-period Moving Average', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Monthly milk production (pounds per cow)')
plt.title('Original Data Series with Moving Averages (data_2)')
plt.legend()
plt.grid(True)
plt.show()
```

Based on above graph this seems that the time series is likely additive.

```python


# Decompose the time series into trend, seasonality, and residual components
result_multiplicative = seasonal_decompose(data['Monthly milk production (pounds per cow)'], model='multiplicative', period=12)

# Plot the multiplicative decomposition
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(data['Monthly milk production (pounds per cow)'], label='Original Data')
plt.legend()
plt.title('Original Data')

plt.subplot(412)
plt.plot(result_multiplicative.trend, label='Trend (Multiplicative)')
plt.legend()

plt.subplot(413)
plt.plot(result_multiplicative.seasonal, label='Seasonal (Multiplicative)')
plt.legend()

plt.subplot(414)
plt.plot(result_multiplicative.resid, label='Residual (Multiplicative)')
plt.legend()

plt.tight_layout()
plt.show()
```
