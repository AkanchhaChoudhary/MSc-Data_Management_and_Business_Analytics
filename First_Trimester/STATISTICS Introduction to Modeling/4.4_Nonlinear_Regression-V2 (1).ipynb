## Nonlinear Models


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('data_session_4.xlsx', sheet_name='non_lin_01')
df.head()
```

```python
correlation_coefficient = np.corrcoef(df['X'], df['Y'])[0, 1]
print("Correlation coefficient:", correlation_coefficient)
```

```python
X = df.X.values
Y = df.Y.values     
plt.plot(X,Y,'*')
```

```python
X.shape
```

```python
X= X.reshape(-1,1)
X.shape
```

```python
X_expanded = np.concatenate([X, X**2], axis=1)
```

```python
mymodel=LinearRegression().fit(X_expanded, Y)
```

```python
print(mymodel.coef_)
print(mymodel.intercept_)
print(mymodel.score(X_expanded, Y))
```

```python
score = mymodel.score(X_expanded, Y)
print(score)
```

So the model is:

$Y = 3.1 X^2 - 5.4 X +146 + $noise,  $R^2 = 0.91$

```python
Y_pred = mymodel.predict(X_expanded)

plt.plot(X,Y,'*')
plt.plot(X,Y_pred, '-g')
plt.show()
```
